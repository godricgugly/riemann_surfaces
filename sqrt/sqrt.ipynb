```python
import numpy as np
import cmath
import plotly.graph_objects as go
```

```python
def build_surface(X, Y, W_list,
                  x_axis="real_z",
                  y_axis="imag_z",
                  z_axis="real_w",
                  color_axis=None,
                  colorscale="HSV",
                  opacity=0.8):
    
    fig = go.Figure()
    
    for W in W_list:
        
        mapping = {
            "real_z": X,
            "imag_z": Y,
            "real_w": np.real(W),
            "imag_w": np.imag(W),
            "abs_w": np.abs(W),
            "arg_w": np.angle(W)
        }
        
        X_plot = mapping[x_axis]
        Y_plot = mapping[y_axis]
        Z_plot = mapping[z_axis]
        
        if color_axis is None:
            fig.add_surface(
                x=X_plot,
                y=Y_plot,
                z=Z_plot,
                colorscale=[[0, "lightgray"], [1, "lightgray"]],
                showscale=False,
                opacity=opacity
            )
        else:
            fig.add_surface(
                x=X_plot,
                y=Y_plot,
                z=Z_plot,
                surfacecolor=mapping[color_axis],
                colorscale=colorscale,
                showscale=True,
                opacity=opacity
            )
    
    fig.update_layout(
        title="Full Riemann Surface",
        scene=dict(
            xaxis_title=x_axis,
            yaxis_title=y_axis,
            zaxis_title=z_axis
        )
    )
    
    fig.show()
```

```python
x = np.linspace(-2, 2, 200)
y = np.linspace(-2, 2, 200)
X, Y = np.meshgrid(x, y)
Z = X + 1j*Y

W1 = np.sqrt(Z)
W2 = -np.sqrt(Z)
```

```python
build_surface(
    X, Y,
    W_list=[W1, W2],
    x_axis="real_z",
    y_axis="imag_z",
    z_axis="real_w",
    color_axis="imag_w"
)
```

```python
build_surface(
    X, Y,
    W_list=[W1, W2],
    x_axis="imag_z",
    y_axis="real_w",
    z_axis="imag_w",
    color_axis="real_z"
)
```
