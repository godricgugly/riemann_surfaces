```python
import numpy as np
import plotly.graph_objects as go
import cmath
```

```python
def plot_riemann_surface_with_point(X, Y, W_list, z0,
                                    x_axis="real_z",
                                    y_axis="imag_z",
                                    z_axis="real_w",
                                    color_axis="imag_w",
                                    point_size=8):
    """
    Plot Riemann surface sheets with 3 spatial axes + 1 color axis.
    Overlays z0 (in base plane) and w1/w2 points on the surface.
    
    Parameters:
        X, Y: meshgrid for z-plane
        W_list: list of complex surfaces (each sheet)
        z0: single complex number to highlight
        x_axis, y_axis, z_axis, color_axis: choice of which components map to which axes
    """
    fig = go.Figure()
    
    # Helper mapping
    def get_mapping(X, Y, W):
        return {
            "real_z": X,
            "imag_z": Y,
            "real_w": np.real(W),
            "imag_w": np.imag(W),
            "abs_w": np.abs(W),
            "arg_w": np.angle(W)
        }
    
    # --- Draw Riemann surface sheets ---
    for W in W_list:
        mapping = get_mapping(X, Y, W)
        fig.add_surface(
            x=mapping[x_axis],
            y=mapping[y_axis],
            z=mapping[z_axis],
            surfacecolor=mapping[color_axis],
            colorscale='HSV',
            showscale=True,
            opacity=0.85
        )
    
    # --- Draw base plane for z0 reference (gray) ---
    fig.add_surface(
        x=mapping["real_z"],
        y=mapping["imag_z"],
        z=np.zeros_like(X),
        colorscale=[[0, "lightgray"], [1, "lightgray"]],
        showscale=False,
        opacity=0.3
    )
    
    # --- Draw z0 on base plane ---
    fig.add_trace(go.Scatter3d(
        x=[np.real(z0)],
        y=[np.imag(z0)],
        z=[0],
        mode='markers',
        marker=dict(size=point_size-6, color='black', symbol='x'),
        name=f"z0 = {z0}"
    ))
    
    # --- Compute w1 and w2 ---
    W0_list = [cmath.sqrt(z0), -cmath.sqrt(z0)]
    
    for i, W0 in enumerate(W0_list):
        mapping_point = {
            "real_z": np.real(z0),
            "imag_z": np.imag(z0),
            "real_w": np.real(W0),
            "imag_w": np.imag(W0),
            "abs_w": np.abs(W0),
            "arg_w": np.angle(W0)
        }
        # Marker on surface
        fig.add_trace(go.Scatter3d(
            x=[mapping_point[x_axis]],
            y=[mapping_point[y_axis]],
            z=[mapping_point[z_axis]],
            mode='markers',
            marker=dict(
                size=point_size,
                color=mapping_point[color_axis],
                colorscale='HSV',
                symbol='circle',
                line=dict(width=2, color='black')
            ),
            name=f"w{i+1} = {W0}"
        ))
        # Optional line connecting z0 → w_i
        fig.add_trace(go.Scatter3d(
            x=[np.real(z0), mapping_point[x_axis]],
            y=[np.imag(z0), mapping_point[y_axis]],
            z=[0, mapping_point[z_axis]],
            mode='lines',
            line=dict(color='black', width=2, dash='dash'),
            showlegend=False
        ))
    
    fig.update_layout(
        title=f"Riemann Surface with z0={z0}",
        scene=dict(
            xaxis_title=x_axis,
            yaxis_title=y_axis,
            zaxis_title=z_axis
        )
    )
    
    fig.show()
```

```python
# Build meshgrid
x = np.linspace(-2, 2, 300)
y = np.linspace(-2, 2, 300)
X, Y = np.meshgrid(x, y)
Z = X + 1j*Y

# Riemann surface sheets
W1 = np.sqrt(Z)
W2 = -np.sqrt(Z)

# Input point
z0 = -1 + 1j

# Choose mapping for 3D + color
plot_riemann_surface_with_point(
    X, Y,
    W_list=[W1, W2],
    z0=z0,
    x_axis="real_z",
    y_axis="imag_z",
    z_axis="real_w",
    color_axis="imag_w"
)
```
