```python
import numpy as np
import cmath
import plotly.graph_objects as go
```

```python
def build_surface(X, Y,
                  f=None,
                  W_list=None,
                  n_branches=1,
                  x_axis="real_z",
                  y_axis="imag_z",
                  z_axis="real_w",
                  color_axis=None,
                  colorscale="HSV",
                  opacity=0.8):
    
    Z = X + 1j * Y
    fig = go.Figure()
    
    # --- Generate W_list automatically if function is given ---
    if W_list is None and f is not None:
    
        r = np.abs(Z)
        theta = np.angle(Z)
    
        W_list = []
    
        if n_branches == 1:
            W_list = [f(Z)]
    
        else:
            for k in range(n_branches):
                Wk = r**(1/n_branches) * np.exp(1j * (theta + 2*np.pi*k) / n_branches)
                W_list.append(Wk)
    
    elif W_list is None:
        raise ValueError("Provide either f or W_list")
    
    # --- Plot each sheet ---
    for W in W_list:
        
        mapping = {
            "real_z": X,
            "imag_z": Y,
            "real_w": np.real(W),
            "imag_w": np.imag(W),
            "abs_w": np.abs(W),
            "arg_w": np.angle(W)
        }
        
        X_plot = mapping[x_axis]
        Y_plot = mapping[y_axis]
        Z_plot = mapping[z_axis]
        
        if color_axis is None:
            fig.add_surface(
                x=X_plot,
                y=Y_plot,
                z=Z_plot,
                colorscale=[[0, "lightgray"], [1, "lightgray"]],
                showscale=False,
                opacity=opacity
            )
        else:
            fig.add_surface(
                x=X_plot,
                y=Y_plot,
                z=Z_plot,
                surfacecolor=mapping[color_axis],
                colorscale=colorscale,
                showscale=True,
                opacity=opacity
            )
    
    fig.update_layout(
        title="Full Riemann Surface",
        scene=dict(
            xaxis_title=x_axis,
            yaxis_title=y_axis,
            zaxis_title=z_axis
        )
    )
    
    fig.show()
```

```python
build_surface(X, Y, f=np.sqrt, n_branches=2) #No color dimension
```

```python
build_surface(X, Y, f=lambda z: z**(1/3), n_branches=3, color_axis="imag_w") 
```
